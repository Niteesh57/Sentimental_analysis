### Label Encoding the Sentiment Column

### Before LabelEncoder

### After LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['Sentiment']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i]).astype(int)
```
